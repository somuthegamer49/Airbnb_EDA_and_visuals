```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Starting Data Cleaning

```python
df.isnull().sum()
```

```python
df['reviews_per_month'].fillna(0,inplace=True)
```

```python
df['name'].dropna(inplace=True)
df.isnull().sum()
```

```python
df['last_review']=pd.to_datetime(df['last_review'],errors="coerce")
```

```python
df['name'].fillna('No Name',inplace=True)
```

```python
df['host_name'].fillna('No Host Name',inplace=True)
```

```python
df.isna().sum()
```

```python
df['id']=df['id'].astype('str')
df['host_id']=df['host_id'].astype('str')
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

## Data Cleaned

```python
price_neighbourhood_group=df.groupby(df['neighbourhood_group'])['price'].mean().plot(kind='bar')
```

## Manhattan has highest price of hotels followed by Brooklyn and Staten Island

```python
plt.hist(df['price'],bins=50,color='purple')
plt.show()
```

## Most of the hotels have 0 availability 

```python
df['room_type'].value_counts().plot(kind='pie',autopct='%.2f')
```

## Distribution of Room Type in AirBnB hotels

```python
plt.figure(figsize=(10,8))
plt.scatter(df['latitude'],df['longitude'],c=df['price'],cmap='coolwarm',alpha=0.3)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.legend()
plt.show()
```

## Location based price distribution based on Longitude and Latitude

```python
price_neighbourhood=df.groupby(df['neighbourhood'])['price'].mean().sort_values().head(10).plot(kind='bar')
```

```python
corr_matrix=df.corr()
corr_matrix
```

```python
heatmap=plt.imshow(corr_matrix,vmin=-1,vmax=1)
plt.colorbar(heatmap)
```

## number of reviews and reviews per month have positive correlation

```python
plt.figure(figsize=(10,8))
plt.scatter(df['number_of_reviews'],df['price'],alpha=0.3)
plt.xlabel("No.Of Reviews")
plt.ylabel("Price")
plt.title('Price vs No of Reviews')
plt.show()
```
